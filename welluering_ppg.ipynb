```python
import plotly.express as px
import struct

import pandas as pd
import numpy as np
from scipy import signal
  
onehertz = {'spo2':[],
        'hr':[],
        'battery':[],
        'activity':[]
       }
sixtyhertz = []

with open("O2Ring 0543_2022-12-27,13:29:06.rt", "r") as f:
    text = f.read().splitlines()
for idx, line in enumerate(text):
    try:
        line = line.zfill(301)[27:]
        spo2, hr, battery, activity, _, ppg_bytes = struct.unpack("<2BxBx2B5x125s", bytes.fromhex(line))
        ppg = [x for x in ppg_bytes]
        onehertz['spo2'].append(spo2)
        onehertz['hr'].append(hr)
        onehertz['battery'].append(battery)
        onehertz['activity'].append(activity)
        sixtyhertz.extend(ppg)
    except ValueError:
        print(idx, line)
        break

    

  
df1 = pd.DataFrame(onehertz)
df2 = pd.Series(sixtyhertz)

# df1['ppg'] = df2

```

```python
df1
```

```python
sig = df2



sos = signal.butter(2, [0.05, 9], 'bp', fs=125, output='sos')
filtered = signal.sosfilt(sos, sig) #* -1

fig = px.line(df1)
fig.show()
```

```python
fig = px.line(y=[sig,filtered])

fig.show()
```

```python
# Python example - Fourier transform using numpy.fft method

import numpy as np

import matplotlib.pyplot as plotter

 

# How many time points are needed i,e., Sampling Frequency

samplingFrequency   = 125;

amplitude = filtered[0:200000]
 
# Frequency domain representation

fourierTransform = np.fft.fft(amplitude)/len(amplitude)           # Normalize amplitude
fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclude sampling frequency
tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/samplingFrequency
frequencies = values/timePeriod



```

```python
px.line( x=frequencies, y=abs(fourierTransform))
```
